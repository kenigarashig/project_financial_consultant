```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/Nikkei_225"

# Fetch the page content
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Locate the sections with company data
sections = soup.find_all("ul")  # Most of the data is in <ul> tags

nikkei_tickers = []

# Iterate through the sections and extract company names and tickers
for section in sections:
    items = section.find_all("li")  # Each company is listed in <li>
    for item in items:
        text = item.get_text()
        if "TYO:" in text:  # Filter lines that contain tickers
            nikkei_tickers.append(text)
```

```python
import re
tickers_only = []

for item in nikkei_tickers:
    match = re.search(r"TYO: (\d+)", item)  # Regex to find "TYO: 1234"
    if match:
        tickers_only.append(match.group(1))  # Extract only the ticker number
```

```python
nikkei_225=[]

for ticker_symbol in tickers_only:
    print(ticker_symbol)
    ticker = yf.Ticker(ticker_symbol + ".T")
    company_name = ticker.info.get("longName", "N/A")  # Get full company name
    nikkei_225.append({"Index": "Nikkei 225", "Ticker": ticker_symbol, "Company Name": company_name})
```

```python
df_table_nikkei_225=pd.DataFrame(nikkei_225)
df_table_nikkei_225["Ticker"]=df_table_nikkei_225["Ticker"]+".T"
```

```python
currency_tickers=["JPY=X","HKD=X","INR=X"]
exchange_rate_data=yf.Tickers(currency_tickers)
```

```python
exchange_rates={"JPY":exchange_rate_data.tickers["JPY=X"].info.get("previousClose"),
                "HKD":exchange_rate_data.tickers["HKD=X"].info.get("previousClose"),
                "INR":exchange_rate_data.tickers["INR=X"].info.get("previousClose")
                }
```

```python
exchange_rates
```

```python
def convert_to_usd(row):
    if row["Currency"] in exchange_rates and row["Market Cap"]!="N/A":
        return row["Market Cap"]/exchange_rates[row["Currency"]]
    return row["Market Cap"]
```

```python
df_screener["Market Cap USD"]=df_screener.apply(convert_to_usd,axis=1)
```

```python
def rank_stocks (df,column):
    df["Rank"]=df[column].rank(ascending=False,method="min")
    ranked_df=df.sort_values("Rank").head(25)
    return ranked_df[["Rank","Ticker","Company Name",column]]
```
