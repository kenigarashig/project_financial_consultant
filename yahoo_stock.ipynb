```python
for ticker in empresas:
    empresa=yf.Ticker(ticker)
    precios_historicos=empresa.history(period="4y",interval="1d")
    if precios_historicos.empty:
        empresas.remove(ticker)
```

```python
resultado_porcentaje_crecimiento=[]
for ticker in empresas:
    empresa=yf.Ticker(ticker)
    precios_historicos=empresa.history(period="4y",interval="1d")
    fecha_inicio=precios_historicos.index[0]
    fecha_final=precios_historicos.index[-1]
    precio_inicio = precios_historicos.loc[fecha_inicio, "Close"]
    precio_final = precios_historicos.loc[fecha_final, "Close"]
    cambio_porcentual=((precio_final-precio_inicio)/precio_inicio)*100
    resultado_porcentaje_crecimiento.append({"Company":ticker,"cambio %":cambio_porcentual})
    
```

```python
for ticker_symbol in empresas:
    empresa=yf.Ticker(ticker_symbol)
    income_statement_loop=empresa.financials.transpose().reset_index()
    income_statement_loop=income_statement_loop.rename(columns={"index":"Date"})
    income_statement_loop["Company"]=ticker_symbol
    income_statement.append(income_statement_loop)
```

```python
for ticker_symbol in empresas:
    empresa=yf.Ticker(ticker_symbol)
    cash_flow_loop=empresa.cash_flow.transpose().reset_index()
    cash_flow_loop=cash_flow_loop.rename(columns={"index":"Date"})
    cash_flow_loop["Company"]=ticker_symbol
    cash_flow.append(cash_flow_loop)
```

```python
for ticker_symbol in empresas:
    empresa=yf.Ticker(ticker_symbol)
    balance_sheet=empresa.balance_sheet.transpose().reset_index()
    balance_sheet=balance_sheet.rename(columns={"index":"Date"})
    balance_sheet["Company"]=ticker_symbol
    balance_sheets.append(balance_sheet)
```

```python
df_balance_sheets=pd.concat(balance_sheets,axis=0,ignore_index=True)
df_income_statement=pd.concat(income_statement,axis=0,ignore_index=True)
df_cash_flow=pd.concat(cash_flow,axis=0,ignore_index=True)
```

```python
def input_transformation(ticker_symbols):
    columns_used = joblib.load("columns_used.pkl")
    balance_sheets_input=[]
    income_statement_input=[]
    cash_flow_input=[]
    empresa=yf.Ticker(ticker_symbol)

    # Process balance sheet
    bs=empresa.balance_sheet.transpose().reset_index()
    bs=bs.rename(columns={"index":"Date1"})
    bs["Company"]=ticker_symbol
    balance_sheets_input.append(bs)

    # Process income statement
    ist=empresa.financials.transpose().reset_index()
    ist=ist.rename(columns={"index":"Date2"})
    ist["Company"]=ticker_symbol
    income_statement_input.append(ist)

    # Process cash flow
    cf=empresa.cash_flow.transpose().reset_index()
    cf=cf.rename(columns={"index":"Date3"})
    cf["Company"]=ticker_symbol
    cash_flow_input.append(cf)


    df_balance_sheets=pd.concat(balance_sheets_input,axis=0,ignore_index=True)
    df_cash_flow=pd.concat(income_statement_input,axis=0,ignore_index=True)
    df_income_statement=pd.concat(cash_flow_input,axis=0,ignore_index=True)
    
    # Merge data on the 'Company' column
    df_all_input=pd.merge(df_balance_sheets,df_income_statement,on="Company",how="inner")
    df_all_input=pd.merge(df_all_input,df_cash_flow,on="Company",how="inner")
    # df_all_input.drop(columns=["Date1","Date2","Date3"], axis=1, inplace=True)
    df_all_input=df_all_input.select_dtypes(include="number")

    for col in columns_used:
        if col not in df_all_input:
            df_all_input[col] = -1

    df_all_input=df_all_input.fillna(-1)

    return df_all_input
```
